```python
import pandas as pd
import os
import glob
import numpy as np
```

```python
input_path = os.getcwd() 
print(input_path)
#현재 스크립트가 들어 있는 폴더의 위치를 알아낸다. 
```

```python
all_files = glob.glob(os.path.join(input_path, '*.txt'))
#현재 폴더에서 확장자명이 txt인 파일의 경로를 모두 추출한다. 
print(all_files)
print(type(all_files))
```

```python
all_data_frames = []
# 출결 파일에서 받아들인 데이터들을 임시로 리스트 형태로 저장하는 변수임
```

```python
for file in all_files:
    data_frame = pd.read_csv(file, index_col=None, encoding="CP949")
    # 윈도우 인코딩 문제를 해결하기 위해서 cp949 설정을 집어넣었음
    base = os.path.basename(file) # 경로에서 파일명만 추출   
    name = os.path.splitext(base)[0] # 파일명에서 파싱을 통해서 확장자를 제거
    data_frame['적용범위'] = name # 파일명을 날짜로 설정했기 때문에 컬럼에 name을 지정함
    del data_frame['Unnamed: 11']
    del data_frame['0']
    # 필요없는 열을 삭제함
    all_data_frames.append(data_frame)
```

```python
print(type(all_data_frames))
```

```python
df = pd.concat(all_data_frames, ignore_index=True)
```

```python
df = df.rename(columns = {'적용범위': '날짜'})
```

```python
df
```

```python
df = df.dropna(axis=0, subset=['이름'])
#axis = 0 이면 nan값 있는 행 삭제
#axis = 1 이면 nan값 있는 열 삭제
```

```python
df
```

```python
df['출결사항'] = np.nan
```

```python
df.index = range(len(df.index))
#인덱스를 새롭게 설정했다. 
```

```python
for index, row in df.iterrows():
    temp = []
    for i in range(3,10):
        # print(df.ix[index,i])
        temp.append(df.ix[index,i]) # 행렬을 이용해서 데이터프레임의 값을 찾아들어갔다. 
    temp = list(set(temp)) # 중복값을 제거한다. 
    if "·" in temp:
        temp.remove("·")
    # "·" 을 제거한다. 
    if not temp:
        temp.append("출석")
    # 리스트가 비어있다면 nan 값을 집어넣는다. 
    print(temp[0])
    df.ix[index,10] = temp[0]
    
        
  
    
    
```
